```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
SANGO = pd.read_csv('SANGO.txt', index_col=None) # SANGO LAT - 7.4232978 and LONG - 3.8984295.
SANGO.set_index(['DAY'], inplace=True)
SANGO.index.names = ['DAY']
SANGO.columns.names = ['TIME']
SANGO.plot.bar()
plt.annotate('The time(6-9pm)with the lowest\n Concentration of Carbon monoxide',
xy=(0,133), xytext=(0,120), arrowprops=dict(facecolor='black',shrink=0.05));
plt.annotate('The time(12-3pm)with the highest \n Concentration of Carbon monoxide',
xy=(5,451), xytext=(0,290), arrowprops=dict(facecolor='red',shrink=0.05))
plt.title('CARBON MONOXIDE CONCENTRATION AT SANGO ON \n 18TH – 24TH NOV, 2019 WITH LAT - 7.4232978 and LONG - 3.8984295')
plt.ylabel('CARBON MONOXIDE CONCENTRATION')
plt.savefig('SANGO3')
plt.show()
```
