```python
import pandas as pd
import numpy as np
A = pd.read_csv('WAEC.RESULT.txt')
A[:5]
```

```python
A.Sex.unique()
```

```python
A.Sex = A.Sex.map({"Male": 1,"Female": 0})
A[:5]
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
model = smf.logit("Sex ~ Maths_Scores + English_Scores", data=A)
result = model.fit()
print(result.summary())
```

```python
print(result.get_margeff().summary())
```

```python
A_scores = pd.DataFrame({"Maths_Scores": np.arange(40,81,10), "English_Scores": np.arange(55,78,5)})
A_scores
```

```python
A_scores["Subjects"] = result.predict(A_scores)
A_scores["Subjects"]
```

```python
Sex=(A_scores["Subjects"] > 0.5).astype(int)
Sex
```

```python
params = result.params
params
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
# sex Female
ax.plot(A[A.Sex == 0].Maths_Scores.values, A[A.Sex == 0].English_Scores.values, 's', label='Female')
ax.plot(A_scores[Sex == 0].Maths_Scores.values, A_scores[Sex == 0].English_Scores.values,
'o', markersize=10, color="steelblue", label='Female(pred.)')
# sex Male
ax.plot(A[A.Sex == 1].Maths_Scores.values, A[A.Sex == 1].English_Scores.values,'s', label='Male')
ax.plot(A_scores[Sex == 1].Maths_Scores.values, A_scores[Sex == 1].English_Scores.values,
'o', markersize=10, color="green", label='Male(pred.)') 
ax.set_xlabel('Female')
ax.set_ylabel('Male')
ax.legend()
plt.show()
```
